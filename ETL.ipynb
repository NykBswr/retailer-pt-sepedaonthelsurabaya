## Data Nilai Tambah & Pertumbuhan Produksi Tahunan (xlsx)
- Nilai Tambah → https://www.bps.go.id/id/statistics-table/2/NDUxIzI=/nilai-tambah-industri-skala-mikro-dan-kecil-menurut-2-digit-kbli--juta-rupiah-.html
- Pertumbuhan Produksi Tahunan → https://www.bps.go.id/id/statistics-table/2/NDA1IzI=/growth-of-production-indices-y-on-y-by-province.html

```python
import pandas as pd
# Load the Excel files
file1 = f'{path}Nilai Tambah Industri Skala Mikro dan Kecil Menurut Provinsi.xlsx'
file2 = f'{path}Pertumbuhan Produksi Tahunan Y on Y Industri Skala Mikro dan Kecil Menurut Provinsi.xlsx'

# Read Excel sheets into DataFrame
df_nilaitambah = pd.read_excel(file1)
df_pertumbuhan = pd.read_excel(file2)
```

### Nilai Tambah

#### Step 1: Clean column names by stripping extra spaces and fixing any issues

```python
df_nilaitambah.columns = df_nilaitambah.columns.str.strip()
```

#### Step 2: Remove the first two rows, as they are not relevant for data processing

```python
df_nilaitambah_cleaned = df_nilaitambah.iloc[2:].reset_index(drop=True)
```

#### Step 3: Rename '38 pertumbuhan' to 'pertumbuhan' for clarity & deleting missing value

```python
df_nilaitambah_cleaned = df_nilaitambah_cleaned.rename(columns={'KBLI 2 digit (Deskripsi)': 'KBLI'})
df_nilaitambah_cleaned = df_nilaitambah_cleaned.dropna(subset=['KBLI'])
df_nilaitambah_cleaned.head()
```

#### Step 4: Extract 'Kecil', 'Menengah', and 'Besar' columns based on the pattern

```python
df_mikro_nilaitambah = df_nilaitambah_cleaned.iloc[:, :10].copy()
df_kecil_nilaitambah = df_nilaitambah_cleaned.iloc[:, [0, 10, 11, 12, 13, 14, 15, 16, 17, 18]].copy()  # Next columns for 'kecil'

# Rename the year columns for each category
year_columns = ['2013', '2014', '2015', '2017', '2018', '2019', '2020', '2021', '2022']
df_mikro_nilaitambah.columns = ['KBLI'] + year_columns
df_kecil_nilaitambah.columns = ['KBLI'] + year_columns
```

#### Step 5: Add 'Kategori' to each dataframe and melt the data into long format

```python
df_kecil_nilaitambah['Jenis_Skala'] = 'Kecil'
df_mikro_nilaitambah['Jenis_Skala'] = 'Mikro'
df_mikro_nilaitambah.head()
```

#### Final step: Combining dataset and remove any unnecessary row

```python
# Combine the three datasets
df_combined_nilaitambah = pd.concat([df_kecil_nilaitambah, df_mikro_nilaitambah], ignore_index=True)
# Melt the dataframe to long format
df_final_nilaitambah = pd.melt(df_combined_nilaitambah, id_vars=['KBLI', 'Jenis_Skala'], value_vars=year_columns, var_name='Tahun', value_name='Nilai_Tambah_Industri')
# Remove any unnecessary rows in the 'KBLI' column
df_final_nilaitambah = df_final_nilaitambah[~df_final_nilaitambah['KBLI'].str.contains('catatan', case=False, na=False)]
df_final_nilaitambah = df_final_nilaitambah[~df_final_nilaitambah['KBLI'].str.contains('INDONESIA', case=False, na=False)]
df_final_nilaitambah = df_final_nilaitambah[~df_final_nilaitambah['KBLI'].str.contains('JUMLAH', case=False, na=False)]
```

```python
df_final_nilaitambah.head()
```

#### Saving final data

```python
df_final_nilaitambah.to_excel(f'{path}/Fix/nilai_tambah_industri_menurut_kbli.xlsx', index=False)
```

```python
df_perusahaan_provinsi.columns = df_perusahaan_provinsi.columns.str.strip()
df_perusahaan_kbli.columns = df_perusahaan_kbli.columns.str.strip()
```

### Step 2: Remove the first two rows, as they are not relevant for data processing


```python
df_perusahaan_provinsi_cleaned = df_perusahaan_provinsi.iloc[2:].reset_index(drop=True)
df_perusahaan_kbli_cleaned = df_perusahaan_kbli.iloc[2:].reset_index(drop=True)
```

### Step 3: Rename '38 Provinsi' to 'Provinsi' for clarity & deleting missing value


```python
df_perusahaan_provinsi_cleaned = df_perusahaan_provinsi_cleaned.rename(columns={'38 Provinsi': 'Provinsi'})
df_perusahaan_provinsi_cleaned = df_perusahaan_provinsi_cleaned.dropna(subset=['Provinsi'])
df_perusahaan_kbli_cleaned = df_perusahaan_kbli_cleaned.rename(columns={'KBLI 2 digit (Deskripsi)': 'KBLI'})
df_perusahaan_kbli_cleaned = df_perusahaan_kbli_cleaned.dropna(subset=['KBLI'])
```

### Step 4: Extract 'Kecil', 'Menengah', and 'Besar' columns based on the pattern

```python
df_mikro_provinsi = df_perusahaan_provinsi_cleaned.iloc[:, :10].copy()
df_kecil_provinsi = df_perusahaan_provinsi_cleaned.iloc[:, [0, 10, 11, 12, 13, 14, 15, 16, 17, 18]].copy()  # Next columns for 'kecil'
df_mikro_kbli = df_perusahaan_kbli_cleaned.iloc[:, :10].copy()
df_kecil_kbli = df_perusahaan_kbli_cleaned.iloc[:, [0, 10, 11, 12, 13, 14, 15, 16, 17, 18]].copy()  # Next columns for 'kecil'

# Rename the year columns for each category
year_columns = ['2013', '2014', '2015', '2017', '2018', '2019', '2020', '2021', '2022']
df_mikro_provinsi.columns = ['Provinsi'] + year_columns
df_kecil_provinsi.columns = ['Provinsi'] + year_columns
df_mikro_kbli.columns = ['KBLI'] + year_columns
df_kecil_kbli.columns = ['KBLI'] + year_columns
```

## Transform Dataset into dimension and fact

### Creating Dimension Tables based on the provided data structure

#### Dimensi Provinsi

```python
dim_provinsi_df = df_final_provinsi[['Provinsi']].drop_duplicates()
dim_provinsi_df['Provinsi_ID'] = dim_provinsi_df.index + 1
dim_provinsi_df.head()
```

#### Dimensi Sektor Industri (KBLI)

```python
dim_sektor_industri_df = df_final_kbli[['KBLI']].drop_duplicates()
dim_sektor_industri_df['Sektor_ID'] = dim_sektor_industri_df.index + 1
dim_sektor_industri_df['Nama_Sektor'] = dim_sektor_industri_df['KBLI'].apply(lambda x: ' '.join(x.split()[1:]))
dim_sektor_industri_df.head()
```

#### Dimensi Waktu

```python
dim_waktu_df = df_pertumbuhan[['Tahun']].drop_duplicates()
dim_waktu_df.reset_index(drop=True, inplace=True)
dim_waktu_df['ID_Waktu'] = dim_waktu_df.index + 1
dim_waktu_df.head()
```

#### Dimensi Skala Industri

```python
dim_skala_industri_df = df_final_kbli[['Jenis_Skala']].drop_duplicates()
dim_skala_industri_df.reset_index(drop=True, inplace=True)
dim_skala_industri_df['Skala_ID'] = dim_skala_industri_df.index + 1
dim_skala_industri_df.head()
```

### Fact table for "Fakta Jumlah Perusahaan Provinsi"

#### Merging data with Dimensi Provinsi, Dimensi Skala Industri, and Dimensi Waktu

```python
fact_jumlah_perusahaan_provinsi_df = df_final_provinsi.merge(dim_provinsi_df, on='Provinsi', how='left')
fact_jumlah_perusahaan_provinsi_df = fact_jumlah_perusahaan_provinsi_df.merge(dim_waktu_df, on='Tahun', how='left')
fact_jumlah_perusahaan_provinsi_df = fact_jumlah_perusahaan_provinsi_df.merge(dim_skala_industri_df, on='Jenis_Skala', how='left')
fact_jumlah_perusahaan_provinsi_df.head()
```

#### Selecting relevant columns for the fact table

```python
fact_jumlah_perusahaan_provinsi_df = fact_jumlah_perusahaan_provinsi_df[['Provinsi_ID', 'ID_Waktu', 'Skala_ID', 'Jumlah_Perusahaan_Industri']]
fact_jumlah_perusahaan_provinsi_df.head()
```

### Fact table for "Fakta Jumlah Perusahaan KBLI"

#### Merging data with Dimensi Sektor Industri, Dimensi Skala Industri, and Dimensi Waktu

```python
fact_jumlah_perusahaan_kbli_df = df_final_kbli.merge(dim_sektor_industri_df, on='KBLI', how='left')
fact_jumlah_perusahaan_kbli_df = fact_jumlah_perusahaan_kbli_df.merge(dim_waktu_df, on='Tahun', how='left')
fact_jumlah_perusahaan_kbli_df = fact_jumlah_perusahaan_kbli_df.merge(dim_skala_industri_df, on='Jenis_Skala', how='left')
fact_jumlah_perusahaan_kbli_df.head()
```

#### Selecting relevant columns for the fact table

```python
fact_jumlah_perusahaan_kbli_df = fact_jumlah_perusahaan_kbli_df[['KBLI', 'ID_Waktu', 'Skala_ID', 'Jumlah_Perusahaan_Industri']]
fact_jumlah_perusahaan_kbli_df.head()
```

### Fact table for "Fakta Nilai Tambah Industri"

#### Merging data with Dimensi Sektor Industri, Dimensi Skala Idustri, and Dimensi Waktu

```python
fact_nilai_tambah_df = df_final_nilaitambah.merge(dim_sektor_industri_df, on='KBLI', how='left')
fact_nilai_tambah_df = fact_nilai_tambah_df.merge(dim_waktu_df, on='Tahun', how='left')
fact_nilai_tambah_df = fact_nilai_tambah_df.merge(dim_skala_industri_df, on='Jenis_Skala', how='left')
fact_nilai_tambah_df.head()
```

#### Selecting relevant columns for the fact table

```python
fact_nilai_tambah_df = fact_nilai_tambah_df[['Sektor_ID', 'Skala_ID', 'ID_Waktu', 'Nilai_Tambah_Industri']]
fact_nilai_tambah_df.head()
```

### Fact table for "Fakta Pertumbuhan Produksi"

#### Merging data with Dimensi Provinsi and Dimensi Waktu

```python
fact_pertumbuhan_produksi_df = df_pertumbuhan.merge(dim_provinsi_df, on='Provinsi', how='left')
fact_pertumbuhan_produksi_df = fact_pertumbuhan_produksi_df.merge(dim_waktu_df, on='Tahun', how='left')
fact_pertumbuhan_produksi_df.head()
```

#### Selecting relevant columns for the fact table

```python
fact_pertumbuhan_produksi_df = fact_pertumbuhan_produksi_df[['Nilai_Tambah_Industri', 'Provinsi_ID', 'ID_Waktu']]
fact_pertumbuhan_produksi_df.head()
```

### Fact table for "Fakta Nilai Output"

#### Merging data with Dimensi Sektor Industri, Dimensi Skala Industri, and Dimensi Waktu

```python
fact_nilai_output_df = df_final_output.merge(dim_sektor_industri_df, on='KBLI', how='left')
fact_nilai_output_df = fact_nilai_output_df.merge(dim_waktu_df, on='Tahun', how='left')
fact_nilai_output_df = fact_nilai_output_df.merge(dim_skala_industri_df, on='Jenis_Skala', how='left')
fact_nilai_output_df.head()
```

#### Selecting relevant columns for the fact table

```python
fact_nilai_output_df = fact_nilai_output_df[['Nilai_Output_Industri', 'Sektor_ID', 'ID_Waktu', 'Skala_ID']]
fact_nilai_output_df.head()
```
